# The Massachusetts Institute of Technology (MIT)

Francisco Escobar Rivas and GitHub fescobar91

The Massachusetts Institute of Technology (MIT) is a private land-grant research university in Cambridge, Massachusetts. Established in 1861, MIT has since played a key role in the development of modern technology and science. Since it was founded, MIT has been ranked among the top academic institutions in the world.

The original MIT logo was adopted on December 26, 1864, by the Corporation in a design recommended by the Committee on the Seal, a committee established in 1863, with President William Barton Rogers and the Treasurer of the Institute as members.
